# 1. Data Preparation

## 1.1 Read and Clean Data

    Perform following actions on AimoScore and WeakLinks
    Read,
    Print Lenght
    Remove ? and NAN
    Print Lenght
    Print Summary

```python
url = "./Data/Assignment3/AimoScore_WeakLink_big_scores.xls"
AimoScore = pd.read_excel(url)
print(len(AimoScore))
data = AimoScore[~AimoScore.isin(['?'])]
data = data.dropna(axis=0)
AimoScore = pd.DataFrame(data)
print(len(AimoScore))
AimoScore.describe()
```

```python
url = "./Data/Assignment3/20190108 scores_and_weak_links.xlsx"
WeakLinks = pd.read_excel(url, engine='openpyxl')
print(len(WeakLinks))
data = WeakLinks[~WeakLinks.isin(['?'])]
data = data.dropna(axis=0)
WeakLinks = pd.DataFrame(data)
print(len(WeakLinks))
WeakLinks.describe()
```

### <font color='green'>Data was clean. not found any ? and NAN</font>

# 3 Join DataSet

## 3.1 Join DataSet
    Join DataSet using ID
    Check for any NAN values

```python
AimoScoreWeakLinks = pd.merge(AimoScore, WeakLinks,on=["ID"])
AimoScoreWeakLinks.describe()
```

```python
AimoScoreWeakLinks[AimoScoreWeakLinks.isna().any(axis=1)]
```

## 3.2 Find Weakest link
    Finding the weak link using weak link columns(14) that has maximum score in each row and adding it as a new column 'WeakLinks'

```python
AimoScoreWeakLinks["WeakeLinks"] = AimoScoreWeakLinks.loc[:,"ForwardHead":"RightHeelRises"].idxmax(axis=1)
```

```python
cols = AimoScoreWeakLinks.columns
quantative_cols = AimoScoreWeakLinks._get_numeric_data().columns
qualitative_cols = list(set(cols) - set(quantative_cols))

print('\nquantative predictors/variables')
for index, col in enumerate(quantative_cols, start=1):
   print (index, '.\'', col, '\' ', sep='', end='')

print('\n\nqualitative predictors/variables')
for index, col in enumerate(qualitative_cols, start=1):
   print (index, '.\'', col, '\' ', sep='', end='')
```
