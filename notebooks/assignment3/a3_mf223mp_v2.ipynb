# 3 Join DataSet

## 3.1 Join DataSet
    Join DataSet using ID
    Check for any NAN values

```python
AimoScoreWeakLinks = pd.merge(AimoScore, WeakLinks,on=["ID"])
AimoScoreWeakLinks.describe()
```

```python
AimoScoreWeakLinks[AimoScoreWeakLinks.isna().any(axis=1)]
```

## 3.2 Find Weakest link
    Finding the weak link using weak link columns(14) that has maximum score in each row and adding it as a new column 'WeakLinks'

```python
AimoScoreWeakLinks["WeakLinks"] = AimoScoreWeakLinks.loc[:,"ForwardHead":"RightHeelRises"].idxmax(axis=1)
```

```python
cols = AimoScoreWeakLinks.columns
quantative_cols = AimoScoreWeakLinks._get_numeric_data().columns
qualitative_cols = list(set(cols) - set(quantative_cols))

print('\nquantative predictors/variables')
for index, col in enumerate(quantative_cols, start=1):
   print (index, '.\'', col, '\' ', sep='', end='')

print('\n\nqualitative predictors/variables')
for index, col in enumerate(qualitative_cols, start=1):
   print (index, '.\'', col, '\' ', sep='', end='')
```

```python
AimoScoreWeakLinks["WeakLinks"].value_counts()
```

```python
temp = (AimoScoreWeakLinks["WeakLinks"].value_counts()<2)
temp = temp[~temp == False]
AimoScoreWeakLinks = AimoScoreWeakLinks[~AimoScoreWeakLinks["WeakLinks"].isin(temp.index)]
AimoScoreWeakLinks["WeakLinks"].value_counts()
```

## 3.3 Discard extra columns
    Discard the 14 weak link scores from the joint table

```python
AimoScoreWeakLinks.drop(AimoScoreWeakLinks.columns.to_series()["ForwardHead":"RightHeelRises"], axis=1, inplace=True)
len(AimoScoreWeakLinks.columns)
```

```python
AimoScoreWeakLinks.describe()
```

## 4 Classification

### <font color='green'>38 Predictors</font>

Use 80% with 38 predictors (all angle deviations and nasm deviations)

```python
trainingPercent = 0.8

y_train, y_test ,X_train, X_test  = train_test_split(AimoScoreWeakLinks[["WeakLinks"]], AimoScoreWeakLinks.drop(columns=["WeakLinks","AimoScore","SCORE","ID","Date","No_1_Time_Deviation","No_2_Time_Deviation","EstimatedScore"]), train_size = trainingPercent)
```

### Logestic Regression

```python
lr_ovr = LogisticRegression()
lr_ovr.fit(X_train,y_train)
#print(lr_ovr.coef_)
#print(lr_ovr.coef_.shape)
#print(lr_ovr.intercept_)
#print(lr_ovr.intercept_.shape)
lr_ovr_pred = lr_ovr.predict(X_test)
#print(confusion_matrix(y_test,lr_ovr_pred))
print(classification_report(y_test, lr_ovr_pred, digits = 3))
print('Accuracy:' + str(accuracy_score(y_test, lr_ovr_pred)))
```

```python
lr_mn = LogisticRegression()
lr_mn.fit(X_train,y_train)
#print(lr_mn.coef_)
#print(lr_mn.coef_.shape)
#print(lr_mn.intercept_)
#print(lr_mn.intercept_.shape)
lr_mn_pred = lr_mn.predict(X_test)
#print(confusion_matrix(y_test,lr_mn_pred))
print(classification_report(y_test, lr_mn_pred, digits = 3))
print('Accuracy:' + str(accuracy_score(y_test, lr_mn_pred)))
```

### Naive Bayes Classification

```python
gnb = GaussianNB()

gnb_pred = gnb.fit(X_train,y_train).predict(X_test)
#print(confusion_matrix(y_test, gnb_pred))
print('Accuracy:' + str(accuracy_score(y_test, gnb_pred)))
```

```python
mnb = MultinomialNB()

mnb_pred = mnb.fit(X_train,y_train).predict(X_test)
#print(confusion_matrix(y_test, mnb_pred))
print('Accuracy:' + str(accuracy_score(y_test, mnb_pred)))
```

### LDA

```python
lda = LinearDiscriminantAnalysis()
lda_model = lda.fit(X_train, y_train)
#print(lda_model.priors_)
#print(lda_model.means_)
#print(lda_model.coef_)
lda_model_pred = lda_model.predict(X_test)
#print(confusion_matrix(y_test, lda_model_pred))
print(classification_report(y_test, lda_model_pred, digits = 3))
print('Accuracy:' + str(accuracy_score(y_test, lda_model_pred)))
```
